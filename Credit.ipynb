Day - 2

```python
x=data.drop("risk",axis=1)
y=data["risk"]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

#Balancing the data using SMOTE

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42)
x_train_rsm,y_train_rsm=smote.fit_resample(x_train,y_train)
```

Day - 3

Train the model

```python
from sklearn.model_selection import KFold,StratifiedKFold


from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score

rand_s=cross_val_score(RandomForestClassifier(),X=x,y=y,cv=5)
Grad_s=cross_val_score(GradientBoostingClassifier(),X=x,y=y,cv=5)
log_s=cross_val_score(LogisticRegression(),X=x,y=y,cv=5)
xg=cross_val_score(XGBClassifier(),X=x,y=y,cv=5)

svm_s=cross_val_score(SVC(),X=x,y=y,cv=5)
Ada_s=cross_val_score(AdaBoostClassifier(),X=x,y=y,cv=5)
knn_s=cross_val_score(KNeighborsClassifier(),X=x,y=y,cv=5)
Dec_s=cross_val_score(DecisionTreeClassifier(),X=x,y=y,cv=5)
```

```python
print("Random Forest          :",rand_s.mean())
print("GradientBoost          :",Grad_s.mean())
print("LogisticRegression     :",log_s.mean())
print("XGboost                :",xg.mean())
print("XGboost                :",xg.mean())
print("Support Vector Machine :",xg.mean())
print("AdaBoost               :",Ada_s.mean())
print("KNeighbors             :",knn_s.mean())
print("DecisionTree           :",xg.mean())
```

```python
model=LogisticRegression()
model.fit(x_train_rsm,y_train_rsm)

pred_y=model.predict(x_test)
print(model.score(x_test,y_test))
```

```python
from sklearn.metrics import accuracy_score

pred_y = model.predict(x_test)
acc = accuracy_score(y_test, pred_y)
print("Accuracy:", acc)
```
